```python
import os
```

```python
df=pd.read_csv(r"C:\Users\HDC0422088\Downloads\inventory_dataset.csv")
```

```python
df
```

```python
null_columns = df.columns[df.isnull().any()]
```

```python
null_columns
```

```python
purchase_price_median=df['Purchase_Price'].median()
purchase_price_median
```

```python
type(df['Purchase_Price'])
```

```python
df.dtypes
```

```python
df['Purchase_Price']=df['Purchase_Price'].fillna(purchase_price_median)
```

```python
df
```

```python
from sklearn.impute import SimpleImputer
```

```python
impute_mean=SimpleImputer(strategy='mean')
```

```python
Sales_price_mean=df[['Sale_Price']].values
```

```python
df[['Sale_Price']]=impute_mean.fit_transform(Sales_price_mean)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
q1=df['Sale_Price'].quantile(0.25)
q2=df['Sale_Price'].quantile(0.75)
```

```python
IQR=q2-q1
lb=q1-1.5*IQR
ub=q2+1.5*IQR
```

```python
Cleaned_df=df[(df['Sale_Price']<lb) | (df['Sale_Price']>ub)]
```

```python
Cleaned_df
```

```python
min_stock = Cleaned_df['Quantity_In_Stock'].min()
max_stock = Cleaned_df['Quantity_In_Stock'].max()
```

```python
Cleaned_df['Quantity_In_Stock_Normalized'] = (Cleaned_df['Quantity_In_Stock'] - min_stock) / (max_stock - min_stock)
```

```python
Cleaned_df
```

```python
Cleaned_df.loc[:,'Quantity_In_Stock_Normalized']=(Cleaned_df['Quantity_In_Stock_Normalized']*100).astype(int)
```

```python
Cleaned_df
```

```python
Cleaned_df.loc[:,'Discontinued']=(Cleaned_df['Discontinued']*100).astype(int)
```

```python
Cleaned_df
```

```python
Cleaned_df.to_csv('cleaned_inventory.csv' , index=False)
```
